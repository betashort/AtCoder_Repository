## B:Palindrome-philia
### 問題文

高八士君は回文が大好きで、回文でない文字列が許せません。高八士君は文字列を1回ハグするごとに、文字列から1文字を選んで任意の文字に変えることができます。
文字列Sが与えられます。
Sを回文にするために必要なハグの最小回数を答えてください。

文字列をリスト型にする。  
文字リストの逆順リストを作る。  
比較する。

```python
s = list(input())
s_r = s[::-1]

count=0
i=0
while("".join(s) != "".join(s_r)):
    
    if (s[i] != s_r[i]) and i<len(s):
        s[i] = s_r[i]
        s_r = s[::-1]
        count+=1
    i+=1
print(count)
```
